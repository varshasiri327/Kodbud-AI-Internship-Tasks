```python
def chatbot():
    print("Hello! I am a simple rule-based chatbot. You can ask me questions or just say hi.")
    print("Type 'exit' to end our conversation.")

    rules = {
        "hello": "Hi there! How can I help you today?",
        "hi": "Hello! What's on your mind?",
        "how are you": "I'm just a computer program, so I don't have feelings, but I'm ready to assist you!",
        "what is your name": "I am a chatbot created to help you.",
        "what can you do": "I can answer simple questions based on my predefined rules. Try asking me about my name or how I am!",
        "bye": "Goodbye! Have a great day!",
        "thank you": "You're welcome!"
    }

    while True:
        user_input = input("You: ").lower().strip()

        if user_input == 'exit':
            print("Chatbot: Goodbye!")
            break
        elif user_input in rules:
            print(f"Chatbot: {rules[user_input]}")
        else:
            print("Chatbot: I'm sorry, I don't understand that. Can you please rephrase or ask something else?")

chatbot()
```
