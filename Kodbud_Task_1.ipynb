```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-nlp/master/data/sms.tsv"
```

```python
data_url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
df = pd.read_csv(data_url, sep='\t', header=None, names=['label', 'text'])

print(f"Successfully loaded dataset via link! Total rows: {len(df)}")
print(df.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

```python
predictions = model.predict(X_test_tfidf)
print(f"\nReal-world Dataset Accuracy: {accuracy_score(y_test, predictions) * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```

```python
custom_email = ["URGENT! You have won a free holiday cruise to the Bahamas! Click this link to claim your ticket now!"]
custom_tfidf = vectorizer.transform(custom_email)
live_prediction = model.predict(custom_tfidf)
print(f"Live Test Result: The message was classified as -> {live_prediction[0].upper()}")
```
